### Prediction using ML Models

##### Before applying logistic regression, we will have to make the data fit for performing logistic regression.

```python
# Cleaning the data, replacing the null values in numeric data by 0 and object data by unknown,

numcolumn = dfcopy.select_dtypes(include = [np.number]).columns
objcolumn = dfcopy.select_dtypes(include = ['object']).columns
```

```python
# Substituting 0 and unknown,

dfcopy[numcolumn] = dfcopy[numcolumn].fillna(0)
dfcopy[objcolumn] = dfcopy[objcolumn].fillna("unknown")
```

```python
dfcopy.head()
```

```python
#Encoding the data,

def generate_code_mapping():
    # df = df.copy() 
    list =[('infections', 139),
                ('neoplasms', (239 - 139)),
                ('endocrine', (279 - 239)),
                ('blood', (289 - 279)),
                ('mental', (319 - 289)),
                ('nervous', (359 - 319)),
                ('sense', (389 - 359)),
                ('circulatory', (459-389)),
                ('respiratory', (519-459)),
                ('digestive', (579 - 519)),
                ('genitourinary', (629 - 579)),
                ('pregnancy', (679 - 629)),
                ('skin', (709 - 679)),
                ('musculoskeletal', (739 - 709)),
                ('congenital', (759 - 739)),
                ('perinatal', (779 - 759)),
                ('ill-defined', (799 - 779)),
                ('injury', (999 - 799))] # Your code ranges here
    code_mapping = {}
    count = 1
    for name, num in list:
        for _ in range(num):
            code_mapping[str(count)] = name
            count += 1
    return code_mapping

def encode_diagnosis(df, codes):
    def get_category(num):
        if not num or num in ['unknown', '?']:
            return 'unknown'
        elif num.upper()[0] == 'V':
            return 'supplemental'
        elif num.upper()[0] == 'E':
            return 'injury'
        else:
            lookup = num.split('.')[0]
            return codes.get(lookup, 'unknown')
    
    for col in df.columns:
        df.loc[:, col] = df[col].apply(get_category)
    
    return df

codes = generate_code_mapping()
listcol = ['diag_1', 'diag_2', 'diag_3']
dfcopy[listcol] = encode_diagnosis(dfcopy[listcol], codes)
```

## Normalization Of Data

```python
#Normalization of the data,

listnormal = ['time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications',
                     'number_outpatient', 'number_emergency', 'number_inpatient', 'number_diagnoses']

from sklearn.preprocessing import StandardScaler

normal = StandardScaler()

data[listnormal] = normal.fit_transform(data[listnormal])

data.describe()
```

```python
#Let's store readmitted in y and rest of the columns in X,

Y = data['30readmit']
X = data.drop(['30readmit'], axis =1)
X = pd.get_dummies(X)
```

#### Splitting the data into training and vallidation data sets. The training data will contain 80 % of the data and validation will contain remaining 20%

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size = .2, 
                                                random_state = 7, stratify = Y)
```
